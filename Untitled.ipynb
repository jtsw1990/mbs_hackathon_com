## Trying to estimate covid-state employment numbers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv(r"data/Business_establishments_per_block_by_CLUE_industry.csv")
```

We have hyp. that the total employment numbers show the effect of covid to some degree

We can see that small dip in the estb. graphs which we based our problem statement on

```python
df_emp = df.groupby(["Census year", "CLUE small area"]).sum().reset_index()[["Census year", "CLUE small area", "Total establishments in block"]]
```

```python
estb_trend = df_emp.groupby("Census year").sum().reset_index()
```

```python
estb_trend[13:]["Census year"]
```

```python
plt.figure(figsize=(15, 8))
plt.plot(estb_trend[14:]["Census year"], estb_trend["Total establishments in block"][14:])
plt.plot(estb_trend[14:]["Census year"], estb_trend["Total establishments in block"].rolling(window=3).mean()[14:], "--")
plt.xticks(ticks=[2016, 2017, 2018, 2019],labels=["2016", "2017", "2018", "2019"], rotation=90)
plt.title("Number of Establishments Actual vs 3 year-SMA")
plt.xlabel("Year")
plt.ylabel("# of establishments")
plt.show()
```
